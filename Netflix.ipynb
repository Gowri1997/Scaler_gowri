### 1. Data type modification

### 2. Unnesting

cast , director, country and listed in categories are in nested format so we need to unnest them

```python
df['cast'] = df['cast'].str.split(', ')  # Split values by space
df = df.explode('cast')
df = df.reset_index(drop=True)
```

```python
df['listed_in'] = df['listed_in'].str.split(', ')  # Split values by space
df = df.explode('listed_in')
df = df.reset_index(drop=True)
```

```python
df['country'] = df['country'].str.split(', ')  # Split values by space
df = df.explode('country')
df = df.reset_index(drop=True)
```

```python
df['director'] = df['director'].str.split(', ')  # Split values by space
df = df.explode('director')
df = df.reset_index(drop=True)
```

```python
top_listed_in=df.groupby('listed_in')[['title']].agg({'title':'nunique'}).sort_values(by='title',ascending=False)
top_10_listed_in=top_listed_in.rename(columns={'title':'no_of_movies'}).head(10)
top_10_listed_in = top_10_listed_in.index[:10]
top_10_listed_in
```

```python
#different Categories

df1=df.loc[(df["country"].isin(top3_country))&(df["listed_in"].isin(top_10_listed_in))]

df_plot=df1.groupby(["country","listed_in"])['title'].nunique().reset_index().rename(columns={'title':'no_of_movies'})
sn.barplot(data=df_plot,x='country',y='no_of_movies',hue='listed_in')
```

```python
df2=df.loc[(df["country"].isin(top3_country))&(df["rating"].isin(top_10_rating))]
df_plot=df2.groupby(["country","rating"])['title'].nunique().reset_index().rename(columns={'title':'no_of_movies'})

plt.subplot(2,2,1)
df_plot_india=df_plot[df_plot["country"]=='India']
plt.pie(df_plot_india['no_of_movies'],labels =df_plot_india['rating'] ,radius=2,autopct = '%2.1f%%')
plt.title('India')
plt.show()

plt.subplot(2,2,2)
df_plot_us=df_plot[df_plot["country"]=='United States']
plt.pie(df_plot_us['no_of_movies'],labels =df_plot_us['rating'] ,radius=2,autopct = '%2.1f%%')
plt.title('United States')
plt.show()

plt.subplot(2,2,3)
df_plot_uk=df_plot[df_plot["country"]=='United Kingdom']
plt.pie(df_plot_uk['no_of_movies'],labels =df_plot_uk['rating'] ,radius=2,autopct = '%2.1f%%')
plt.title('United Kingdom')
plt.show()

#sn.barplot(data=df_plot,x='country',y='no_of_movies',hue='rating')
```
